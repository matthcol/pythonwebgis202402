## Conditonal and loops

```python
for n in numbers:
    x = n**2 + 1 
    print(f"{n} -> {x}")
print("The end")
```

```python
# a set is iterable
for n in number_set:
    print(n)
```

```python
for letter in department:
    print(letter, end=' ')
print()
```

```python
print("Dictionary:", city_d)
print()
print("Infos (keys):")
for info in city_d:
    print("  -", info)
print()
print("Infos (keys again):")
for info in city_d.keys():
    print("  -", info)
print()
print("Values:")
for value in city_d.values():
    print("  -", value)
print()
print("Infos + values:")
for info, value in city_d.items():
    print(f"  - {info} = {value}")
```

NB: a lot of values have truth values (str, list, None)

```python
for city in cities_df.label:
    if city.startswith('tart'):
        print("City starting with 'tart':", city)
    elif city.endswith('bes'):
        print("City ending with 'bes':", city)
    elif city.startswith('tou') and city.endswith('louse'):
        print("City matching 'tou..louse'", city)
```

```python
for city in cities_df.label:
    match city:
        case "tarbes":
            print('\nJe suis  Tarbes, youhou !')
        case "toulouse":
            print('\nToulouse ville rose')
        case _:
            print('.', end='')
print()
```
